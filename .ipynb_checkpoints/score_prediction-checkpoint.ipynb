```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Load and Visualise the Data
# - Download
# - Load
# - Visualise
# - Normalise
```

```python
X = pd.read_csv('./Linear_X_Train.csv')
y = pd.read_csv('./Linear_Y_Train.csv')
```

```python
plt.scatter(X , y)
plt.style.use('seaborn-v0_8')
plt.xlabel('Time')
plt.ylabel('Performance')
plt.title('TIme vs Performance Graph')

# normalise the data
u = X.mean()
sd = X.std()

X = (X - u)/sd

plt.show()
```

```python
### Linear Regression

X = X.to_numpy()
y = y.to_numpy()    
```

```python
type(X) , type(y)
```

```python
def hypothesis(x , theta):
    return theta[0] + theta[1]*x

def gradient(X , Y , theta):
    m = len(X)
    grad0 = 0.0
    grad1 = 0.0
    
    for i in range(m):
        hypo = hypothesis(X[i] , theta)
        grad0 += (hypo - Y[i])
        grad1 += (hypo - Y[i])*X[i]
    
    grad0/=m
    grad1/=m
    
    return np.array([grad0 , grad1])

def error(X , Y , theta):
    length = len(X)
    total_error = 0.0
    
    for i in range(length):
        hypo = hypothesis(X[i] , theta)
        yval = Y[i]
        
        total_error += (hypo - yval)**2
        
        
    total_error/=length
    
    return total_error
        
    
    


def gradientDescent(X , Y , learningrate = 0.1 , max_steps = 1000):
    
    theta0 = 0.0
    theta1 = 0.0
    
    theta = [theta0 , theta1]
    
    error_list = []
    
#     grad = gradient(X , Y , theta)
    
    for i in range(max_steps):
        grad = gradient(X , Y , theta)
        theta[0] = theta[0] - learningrate*grad[0]
        theta[1] = theta[1] - learningrate*grad[1]
#         theta = [theta0 , theta1]
        error_list.append(error(X , Y , theta))
    
    return np.array(theta) , error_list


theta , error_list = gradientDescent(X , y)

    
    



        
```

```python
plt.plot(error_list)
plt.title('Reduction error over time')
plt.show()
```

```python
### Predictions and Best Line

y_ = hypothesis(X , theta)
print(y_)
```

```python
# Training + Predictions

plt.scatter(X , y , label= 'Real Data')
plt.plot(X , y_, color= 'orange' , label='Prediction(Best Line)')
plt.legend()

plt.show()
```
